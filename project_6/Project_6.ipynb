## 1. Приоритизация гипотез

### 1.1. Подготовка данных

```python
# <импорт необходимых библиотек>
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import matplotlib
import seaborn as sns
from datetime import datetime
from datetime import timedelta
import scipy.stats as stats
```

```python
# <считывание файла hypothesis.csv -> данные о гипотезах>
try:
    hypothesis = pd.read_csv('hypothesis.csv')
except:
    hypothesis = pd.read_csv('/datasets/hypothesis.csv')

pd.options.display.max_colwidth = 150
hypothesis
```

```python
# <преобразование названий столбцов к нижнему регистру>
hypothesis.columns = hypothesis.columns.map(lambda x: x.lower())
hypothesis.columns
```

```python
# <общая информация о таблице>
hypothesis.info()
```

### 1.2. Приоритизация гипотез при помощи фреймворка ICE

ICE SCORE = $\frac{\textrm{Impact * Confidence}}{\textrm{Efforts}}$

```python
# <добавление результатов ICE SCORE>
hypothesis['ICE'] = hypothesis.impact * hypothesis.confidence / hypothesis.efforts
# <гипотезы по убыванию приоритета>
hypothesis[['hypothesis', 'ICE']].sort_values(by='ICE', ascending=False)
```

```python
# <рейтинг гипотез согласно фреймворку ICE>
print('Рейтинг ICE и название гипотезы\n\n')
hypothesis[['hypothesis', 'ICE']].sort_values(by='ICE', ascending=False).apply(lambda x: print('{:.0f} - {}\n'.format(float(x[1]), str(x[0]))), axis=1)
pass
```

```python
# <создание таблицы с рангами гипотез по фреймворку ICE>
hypothesis_ICE = hypothesis[['hypothesis', 'ICE']].sort_values(by='ICE', ascending=False).reset_index(drop=True).reset_index()[['hypothesis', 'ICE', 'index']]
hypothesis_ICE.columns = ['hypothesis', 'ICE', 'ICE_rank']
hypothesis_ICE['ICE_rank'] = hypothesis_ICE['ICE_rank'] + 1
hypothesis_ICE
```

### 1.3. Приоритизация гипотез при помощи фреймворка RICE

ICE SCORE = $\frac{\textrm{Reach * Impact * Confidence}}{\textrm{Efforts}}$

```python
# <добавление результатов RICE SCORE>
hypothesis['RICE'] = hypothesis.reach * hypothesis.impact * hypothesis.confidence / hypothesis.efforts
# <гипотезы по убыванию приоритета>
hypothesis[['hypothesis', 'RICE']].sort_values(by='RICE', ascending=False)
```

```python
# <рейтинг гипотез согласно фреймворку RICE>
print('Рейтинг RICE и название гипотезы\n\n')
hypothesis[['hypothesis', 'RICE']].sort_values(by='RICE', ascending=False).apply(lambda x: print('{:.2f} - {}\n'.format(float(x[1]), str(x[0]))), axis=1)
pass
```

```python
# <создание таблицы с рангами гипотез по фреймворку ICE>
hypothesis_RICE = hypothesis[['hypothesis', 'RICE']].sort_values(by='RICE', ascending=False).reset_index(drop=True).reset_index()[['hypothesis', 'RICE', 'index']]
hypothesis_RICE.columns = ['hypothesis', 'RICE', 'RICE_rank']
hypothesis_RICE['RICE_rank'] = hypothesis_RICE['RICE_rank'] + 1
hypothesis_RICE.loc[3, 'RICE_rank'] = 3
hypothesis_RICE
```

### 1.4. Приоритизация гипотез: сравнение результатов ICE и RICE

Рассмотрим разницу приоритизации при помощи двух фреймворков.

```python
# <результаты ICE и RICE>
ice_and_rice = hypothesis_ICE.merge(hypothesis_RICE, on='hypothesis')
ice_and_rice = ice_and_rice.merge(hypothesis[['hypothesis', 'reach']], on='hypothesis')
ice_and_rice['ICE_plus_RICE_rank'] = ice_and_rice.ICE_rank + ice_and_rice.RICE_rank
ice_and_rice.sort_values(by='ICE_plus_RICE_rank')
```

**Выводы**

В фреймворке *RICE* по сравнению с фреймворком *ICE* добавляется параметр **Reach (охват)**. Данный параметр учитывает, скольких пользователей коснутся вносимые изменения.

Добавление данного параметра в вычисления повлияло на результаты гипотез для фреймворка RICE, однако ТОП-5 гипотез одинаков для обоих фреймворков.

### 2.1. Кумулятивная выручка по группам

```python
# <подготовка групп и дат>
datesGroups = orders[['date', 'group']].drop_duplicates().reset_index(drop=True)
datesGroups.info()
```

```python
# <кумулятивные данные по заказам>
ordersAggregated = datesGroups.apply(
    lambda x: orders[
        np.logical_and(
            orders['date'] <= x['date'], orders['group'] == x['group']
        )
    ].agg(
        {
            'date': 'max',
            'group': 'max',
            'transactionId': pd.Series.nunique,
            'visitorId': pd.Series.nunique,
            'revenue': 'sum',
        }
    ),
    axis=1,
).sort_values(by=['date', 'group'])
ordersAggregated.columns = ['date', 'group', 'orders', 'buyers', 'revenue']
ordersAggregated.head()
```

```python
# <кумулятивные данные по количеству посетителей>
visitorsAggregated = datesGroups.apply(
    lambda x: visitors[
        np.logical_and(
            visitors['date'] <= x['date'], visitors['group'] == x['group']
        )
    ].agg({'date': 'max', 'group': 'max', 'visitors': 'sum'}),
    axis=1,
).sort_values(by=['date', 'group'])
visitorsAggregated.head()
```

```python
# <сводные кумулятивные данные>
cumulativeData = ordersAggregated.merge(visitorsAggregated, on=['date', 'group'])
cumulativeData.head()
```

```python
# <кумулятивная выручка для двух групп>
cumulativeRevenueA = cumulativeData[cumulativeData['group']=='A'][['date','revenue', 'orders']]
cumulativeRevenueB = cumulativeData[cumulativeData['group']=='B'][['date','revenue', 'orders']]
cumulativeRevenueA.head()
```

```python
# <убираем предупреждение при построении графиков>
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
# <графики кумулятивной выручки по группам>
plt.style.use('ggplot')
plt.figure(figsize=(16, 9))
ax = sns.lineplot(cumulativeRevenueA['date'], cumulativeRevenueA['revenue'], label='группа A')
sns.lineplot(cumulativeRevenueB['date'], cumulativeRevenueB['revenue'], label='группа B')
plt.title('Кумулятивная выручка по группам')
plt.xlabel('Дата')
plt.ylabel('Выручка')
plt.xticks(np.append(ax.get_xticks()[:4], np.append([737289.], ax.get_xticks()[5:])))
ax.get_yaxis().set_major_formatter(
matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')));
```

```python
# <максимальные покупки по пользователям для группы B за 19 августа 2019>
orders.query('(date == "2019-08-19") & (group == "B")').groupby('visitorId')['revenue'].max().sort_values(ascending=False)
```

### 2.10. Статистическая значимость различий в конверсии ("сырые" данные)

```python
# <количество посетителей в день для группы А>
visitorsADaily = visitors[visitors['group'] == 'A'][['date', 'visitors']]
visitorsADaily.columns = ['date', 'visitorsPerDateA']
visitorsADaily.head()
```

```python
# <кумулятивное количество посетителей для группы А>
visitorsACummulative = visitorsADaily.apply(
    lambda x: visitorsADaily[visitorsADaily['date'] <= x['date']].agg(
        {'date': 'max', 'visitorsPerDateA': 'sum'}
    ),
    axis=1,
)
visitorsACummulative.columns = ['date', 'visitorsCummulativeA']
visitorsACummulative.head()
```

```python
# <количество посетителей в день для группы B>
visitorsBDaily = visitors[visitors['group'] == 'B'][['date', 'visitors']]
visitorsBDaily.columns = ['date', 'visitorsPerDateB']
visitorsBDaily.head()
```

```python
# <кумулятивное количество посетителей для группы B>
visitorsBCummulative = visitorsBDaily.apply(
    lambda x: visitorsBDaily[visitorsBDaily['date'] <= x['date']].agg(
        {'date': 'max', 'visitorsPerDateB': 'sum'}
    ),
    axis=1,
)
visitorsBCummulative.columns = ['date', 'visitorsCummulativeB']
visitorsBCummulative.head()
```

```python
# <количество заказов и ежедневная выручка для группы А>
ordersADaily = (
    orders[orders['group'] == 'A'][['date', 'transactionId', 'visitorId', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transactionId': pd.Series.nunique, 'revenue': 'sum'})
)
ordersADaily.columns = ['date', 'ordersPerDateA', 'revenuePerDateA']
ordersADaily.head()
```

```python
# <кумулятивное количество заказов и выручка для группы А>
ordersACummulative = ordersADaily.apply(
    lambda x: ordersADaily[ordersADaily['date'] <= x['date']].agg(
        {'date': 'max', 'ordersPerDateA': 'sum', 'revenuePerDateA': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
ordersACummulative.columns = ['date', 'ordersCummulativeA', 'revenueCummulativeA']
ordersACummulative.head()
```

```python
# <количество заказов и ежедневная выручка для группы В>
ordersBDaily = (
    orders[orders['group'] == 'B'][['date', 'transactionId', 'visitorId', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transactionId': pd.Series.nunique, 'revenue': 'sum'})
)
ordersBDaily.columns = ['date', 'ordersPerDateB', 'revenuePerDateB']
ordersBDaily.head()
```

```python
# <кумулятивное количество заказов и выручка для группы B>
ordersBCummulative = ordersBDaily.apply(
    lambda x: ordersBDaily[ordersBDaily['date'] <= x['date']].agg(
        {'date': 'max', 'ordersPerDateB': 'sum', 'revenuePerDateB': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
ordersBCummulative.columns = ['date', 'ordersCummulativeB', 'revenueCummulativeB']
ordersBCummulative.head()
```

```python
# <создание общей таблицы>
data = (
    ordersADaily.merge(
        ordersBDaily, left_on='date', right_on='date', how='left'
    )
    .merge(ordersACummulative, left_on='date', right_on='date', how='left')
    .merge(ordersBCummulative, left_on='date', right_on='date', how='left')
    .merge(visitorsADaily, left_on='date', right_on='date', how='left')
    .merge(visitorsBDaily, left_on='date', right_on='date', how='left')
    .merge(visitorsACummulative, left_on='date', right_on='date', how='left')
    .merge(visitorsBCummulative, left_on='date', right_on='date', how='left')
)
data.head()
```

```python
# <группировка по пользователям и количеству заказов для группы А>
ordersByUsersA = (
    orders[orders['group'] == 'A']
    .groupby('visitorId', as_index=False)
    .agg({'transactionId': pd.Series.nunique})
)
ordersByUsersA.columns = ['visitorId', 'orders']
ordersByUsersA.head()
```

```python
# <группировка по пользователям и количеству заказов для группы B>
ordersByUsersB = (
    orders[orders['group'] == 'B']
    .groupby('visitorId', as_index=False)
    .agg({'transactionId': pd.Series.nunique})
)
ordersByUsersB.columns = ['visitorId', 'orders']
ordersByUsersB.head()
```

```python
# <первый набор данных - группа А>
sampleA = pd.concat([ordersByUsersA['orders'],
                     pd.Series(0, index=np.arange(data['visitorsPerDateA'].sum() - 
                                                  len(ordersByUsersA['orders'])), name='orders')],axis=0)
sampleA.value_counts()
```

```python
# <второй набор данных - группа В>
sampleB = pd.concat([ordersByUsersB['orders'],
                     pd.Series(0, index=np.arange(data['visitorsPerDateB'].sum() - 
                                                  len(ordersByUsersB['orders'])), name='orders')],axis=0)
sampleB.value_counts()
```

Нулевая гипотеза (H0) : *конверсия для пользователей группы В* **равна** *конверсии для пользователей группы А* ("сырые" данные);

Альтернативная гипотеза (H1): *конверсия для пользователей группы В* **не равна** *конверсии для пользователей группы А* ("сырые" данные).

Теперь посчитаем статистическую значимость различия конверсии по результатам месяца теста. Применим тест Манна-Уитни. Значение *альфа* примем равным **0.05**.

Также вычислим относительное различие в конверсии между группами.

```python
print("P-value теста Манна-Уитни по сравнению конверсии между группами равен {0:.4f}".format(stats.mannwhitneyu(sampleA, sampleB)[1]))
print("Относительный прирост конверсии группы В к конверсии группы А равен {0:.2%}".format((data['ordersPerDateB'].sum()/data['visitorsPerDateB'].sum())/(data['ordersPerDateA'].sum()/data['visitorsPerDateA'].sum())-1))
```

**Выводы и предположения**

*P-value теста Манна-Уитни по сравнению конверсии между группами* равен **0.0055**. *Относительный прирост конверсии группы В к конверсии группы А* равен **15.98%**.

Анализ "сырых данных" сообщает, что *в конверсии между группами* **есть статистически значимые различия** (*p-value < alpha*).

### 2.11. Статистическая значимость различий в среднем чеке ("сырые" данные)

Нулевая гипотеза (H0) : *средний чек для пользователей группы В* **равен** *среднему чеку для пользователей группы А* ("сырые" данные);

Альтернативная гипотеза (H1): *средний чек для пользователей группы В* **не равен** *среднему чеку для пользователей группы А* ("сырые" данные).

Посчитаем статистическую значимость различия средних чеков по результатам месяца теста. Применим тест Манна-Уитни. Значение альфа примем равным 0.05.

Также вычислим относительное различие в средних чеках между группами.

```python
print('P-value теста Манна-Уитни по сравнению среднего чека между группами равен {0:.3f}'.format(stats.mannwhitneyu(orders[orders['group']=='A']['revenue'], orders[orders['group']=='B']['revenue'])[1]))
print('Относительное изменение среднего чека группы В к группе А равно {0:.2%}'.format(orders[orders['group']=='B']['revenue'].mean()/orders[orders['group']=='A']['revenue'].mean()-1))
```

**Выводы и предположения**

Анализ "сырых данных" сообщает, что *в конверсии между группами* **нет статистически значимых различий** (*p-value > alpha*). Благодаря использованию непараметрического статистического метода имеющийся выброс несильно сказался на результатах.

### 2.12. Статистическая значимость различий в конверсии ("очищенные" данные)

Аномального количества заказов не обнаружено, создадим только таблицу для аномально дорогих заказов.

```python
# <пользователи с аномально дорогими заказами>
usersWithExpensiveOrders = orders[orders['revenue'] > 30000]['visitorId']
len(usersWithExpensiveOrders)
```

```python
# <первые 5 пользователей с аномально дорогими заказами>
usersWithExpensiveOrders.head()
```

```python
# <набор очищенных данных для группы А>
sampleAFiltered = pd.concat(
    [
        ordersByUsersA[
            np.logical_not(ordersByUsersA['visitorId'].isin(usersWithExpensiveOrders))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)
sampleAFiltered.value_counts()
```

```python
# <набор очищенных данных для группы В>
sampleBFiltered = pd.concat(
    [
        ordersByUsersB[
            np.logical_not(ordersByUsersB['visitorId'].isin(usersWithExpensiveOrders))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)
sampleBFiltered.value_counts()
```

Нулевая гипотеза (H0) : *конверсия для пользователей группы В* **равна** *конверсии для пользователей группы А* ("очищенные" данные);

Альтернативная гипотеза (H1): *конверсия для пользователей группы В* **не равна** *конверсии для пользователей группы А* ("очищенные" данные).

Теперь посчитаем статистическую значимость различия конверсии по "очищенным" данным по результатам месяца теста. Применим тест Манна-Уитни. Значение *альфа* примем равным **0.05**.

Также вычислим относительное различие в конверсии между группами по "очищенным" данным.

```python
print('P-value теста Манна-Уитни по сравнению конверсии между группами по "очищенным" данным равен {0:.5f}'.format(stats.mannwhitneyu(sampleAFiltered, sampleBFiltered)[1]))
print('Относительный прирост конверсии группы В к конверсии группы А по "очищенным" данным равен {0:.2%}'.format(sampleBFiltered.mean()/sampleAFiltered.mean()-1)) 
```

**Выводы и предположения**

*P-value теста Манна-Уитни по сравнению конверсии между группами по "очищенным" данным* равен **0.00665**. *Относительный прирост конверсии группы В к конверсии группы А по "очищенным" данным* равен **16.1%**.

Анализ "очищенных данных" сообщает, что *в конверсии между группами* **есть статистически значимые различия** (*p-value < alpha*).

### 2.13. Статистическая значимость различий в среднем чеке ("очищенные" данные)

Нулевая гипотеза (H0) : *средний чек для пользователей группы В* **равен** *среднему чеку для пользователей группы А* ("очищенные" данные);

Альтернативная гипотеза (H1): *средний чек для пользователей группы В* **не равен** *среднему чеку для пользователей группы А* ("очищенные" данные).

```python
print(
    'P-value теста Манна-Уитни по сравнению среднего чека между группами по "очищенным" данным равен {0:.3f}'.format(
        stats.mannwhitneyu(
            orders[
                np.logical_and(
                    orders['group'] == 'A',
                    np.logical_not(orders['visitorId'].isin(usersWithExpensiveOrders)),
                )
            ]['revenue'],
            orders[
                np.logical_and(
                    orders['group'] == 'B',
                    np.logical_not(orders['visitorId'].isin(usersWithExpensiveOrders)),
                )
            ]['revenue'],
        )[1]
    )
)

print(
    'Относительное изменение среднего чека группы В к группе А по "очищенным" данным равно {0:.2%}'.format(
        orders[
            np.logical_and(
                orders['group'] == 'B',
                np.logical_not(orders['visitorId'].isin(usersWithExpensiveOrders)),
            )
        ]['revenue'].mean()
        / orders[
            np.logical_and(
                orders['group'] == 'A',
                np.logical_not(orders['visitorId'].isin(usersWithExpensiveOrders)),
            )
        ]['revenue'].mean()
        - 1
    )
)
```

**Выводы и предположения**

*P-value теста Манна-Уитни по средним чекам между группами* равен **0.421**. *Относительное измененме конверсии группы В к конверсии группы А* равно **-3.07%** (засчет исключения аномально высокого значения в группе В).

Анализ "очищенных данных" сообщает, что *в конверсии между группами* **нет статистически значимых различий** (*p-value > alpha*).
