## 3. Ответы на вопросы

### 3.1. Количество книг, опубликованных после 1 января 2000 года

```python
# <количество книг, опубликованных после 1 января 2000 года>
query = \
'''
/* вывод количества книг, опубликованных после 1 января 2000 года */
SELECT COUNT(publication_date) AS books_count
FROM books
WHERE publication_date > '2000-01-01'
'''
pd.io.sql.read_sql(query, con = engine) 
```

**Выводы**

- После 1 января 2000 года было опубликовано **819 книг**;
- Книги, выпущенные 1 января 2000 года, в вычислениях не учитывались.

### 3.2. Количество обзоров и средняя оценка книг

```python
# <количество обзоров и средняя оценка книг>
query = \
'''
/* вывод book_id, названия, количества обзоров и среднего рейтинга по книгам;
   для части книг обзоров нет, для них заменили null на 0;
   такую же проверку для рейтинга проводить не стали, так как рейтинг может быть нулевым,
   если же его нет, то будет стоять null */
SELECT b.book_id, b.title,
       CASE WHEN subq1.review_count IS NOT NULL THEN subq1.review_count ELSE 0 END AS review_count, 
       subq2.avg_rating
FROM books b LEFT JOIN (SELECT book_id, COUNT(book_id) AS review_count FROM reviews GROUP BY book_id) subq1
                       ON b.book_id=subq1.book_id
             LEFT JOIN (SELECT book_id, AVG(rating) AS avg_rating FROM ratings GROUP BY book_id) subq2
                       ON b.book_id=subq2.book_id
ORDER BY book_id
'''
reviews_and_rating = pd.io.sql.read_sql(query, con = engine)
reviews_and_rating
```

**Выводы**

- Всего в таблицу присутствует 1000 книг;
- При формировании запроса выяснилось, что у части книг нет обзоров, поэтому была применена конструкция *CASE WHEN* для замены значений *NULL* на *0*.

### 3.3. Издательство, которое выпустило наибольшее число книг толще 50 страниц

```python
# <издательство, которое выпустило наибольшее число книг толще 50 страниц>
query = \
'''
/* подсчет количества книг толще 50 страниц для каждого издательства */
WITH books_over50_counter AS
(SELECT publisher_id, COUNT(publisher_id) AS book_counter
FROM books 
WHERE num_pages > 50
GROUP BY publisher_id)
        
/* соединение таблицы с издательствами и таблицы с количеством книг,
   вывод издательства с максимальным количеством книг толще 50 страниц */
SELECT p.publisher_id, p.publisher, boc.book_counter
FROM publishers p INNER JOIN books_over50_counter boc ON p.publisher_id=boc.publisher_id
WHERE boc.book_counter = (SELECT MAX(book_counter) FROM books_over50_counter)
'''
pd.io.sql.read_sql(query, con = engine) 
```

**Выводы**

- Издательство, выпустившее наибольшее число книг толще 50 страниц - **Penguin Books** -> **42 книги**;
- Чтобы учесть случай, когда несколько издательств выпустило одинаковое число книг, была использована конструкция после *WHERE* в конце запроса вместо ключевого слова *LIMIT*.

### 3.4. Автор с самой высокой средней оценкой книг — учитывать только книги с 50 и более оценками

```python
# <автор с самой высокой средней оценкой книг — учитывать только книги с 50 и более оценками>
# <расчет среднего среди всех книг без группировки по книгам>
query = \
'''
/* выражение для определения книг с количеством оценок >= 50 */
WITH ratings_over50 AS
(SELECT book_id, rating
FROM ratings
WHERE book_id IN (SELECT book_id FROM ratings GROUP BY book_id HAVING COUNT(book_id) >= 50)),
/* выражение для подсчета средней оценки по автору */
avg_author_ratings as
(SELECT b.author_id, AVG(ro50.rating) AS avg_rating
FROM books b INNER JOIN ratings_over50 ro50 ON b.book_id=ro50.book_id
GROUP BY b.author_id)
        
/* соединение с таблицей authors, выбор автора с максимальной оценкой книг */
SELECT a.author, aar.avg_rating
FROM avg_author_ratings aar INNER JOIN authors a ON aar.author_id=a.author_id
WHERE aar.avg_rating = (SELECT MAX(avg_rating) FROM avg_author_ratings)
'''
pd.io.sql.read_sql(query, con = engine)
```

```python
# <автор с самой высокой средней оценкой книг — учитывать только книги с 50 и более оценками>
# <расчет среднего среди средних оценок по книгам>
query = \
'''
/* выражение для определения средней оценки книг с количеством оценок >= 50 */
WITH ratings_over50 AS
(SELECT book_id, AVG(rating) AS book_avg_rating
FROM ratings
WHERE book_id IN (SELECT book_id FROM ratings GROUP BY book_id HAVING COUNT(book_id) >= 50)
GROUP BY book_id),
        
/* выражение для подсчета средней оценки по автору */
avg_author_ratings AS
(SELECT b.author_id, AVG(ro50.book_avg_rating) AS avg_rating
FROM books b INNER JOIN ratings_over50 ro50 ON b.book_id=ro50.book_id
GROUP BY b.author_id)
        
/* соединение с таблицей authors, выбор автора с максимальной оценкой книг */
SELECT a.author, aar.avg_rating
FROM avg_author_ratings aar INNER JOIN authors a ON aar.author_id=a.author_id
WHERE aar.avg_rating = (SELECT MAX(avg_rating) FROM avg_author_ratings)
'''
pd.io.sql.read_sql(query, con = engine)
```

**Выводы**

- Авторы с самой высокой оценкой книг (при расчете среднего среди всех оценок без группировки по книгам) - **J.K. Rowling/Mary GrandPré** -> средняя оценка составляет **4.287** из 5;
- Авторы с самой высокой оценкой книг (при расчете среднего среди средних оценок книг) - также **J.K. Rowling/Mary GrandPré** -> средняя оценка чуть ниже и составляет **4.284** из 5.

### 3.5. Среднее количество обзоров от пользователей, которые поставили больше 50 оценок

```python
# <среднее количество обзоров от пользователей, которые поставили больше 50 оценок>
query = \
'''
/* выбор пользователей, которые поставили более 50 оценок */
WITH users_over50_ratings AS
(SELECT username
FROM ratings
GROUP BY username
HAVING count(username) > 50),
/* подсчет количества обзоров от пользователей, которые поставили больше 50 оценок */
users_over50_ratings_reviews AS
(SELECT username, COUNT(review_id) AS cnt_reviews
FROM reviews
WHERE username IN (SELECT username FROM users_over50_ratings)
GROUP BY username)

/* вывод среднего количества обзоров от пользователей, которые поставили более 50 оценок */
SELECT AVG(cnt_reviews) AS avg_cnt_reviews
FROM users_over50_ratings_reviews
'''
pd.io.sql.read_sql(query, con = engine)
```

**Выводы**

- В среднем пользователи, которые поставили более 50 оценок, также делают около **24.3 обзоров**.
