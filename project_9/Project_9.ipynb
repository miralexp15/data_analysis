**Целевая переменная:**

- **churn** &mdash; отток клиента в текущем месяце (клиент попал в отток, если за последний месяц ни разу не посетил спортзал).

**Имеющиеся поля:**

- Данные клиента за предыдущий до проверки факта оттока месяц:
    - **gender** — пол;
    - **Near_Location** — проживание или работа в районе, где находится фитнес-центр;
    - **Partner** — сотрудник компании-партнёра клуба (сотрудничество с компаниями, чьи сотрудники могут получать скидки на абонемент — в таком случае фитнес-центр хранит информацию о работодателе клиента);
    - **Promo_friends** — факт первоначальной записи в рамках акции «приведи друга» (использовал промо-код от знакомого при оплате первого абонемента);
    - **Phone** — наличие контактного телефона;
    - **Age** — возраст;
    - **Lifetime** — время с момента первого обращения в фитнес-центр *за месяц до проверки факта оттока* (в месяцах).
- Информация на основе журнала посещений, покупок и информация о текущем статусе абонемента клиента:
    - **Contract_period** — длительность текущего действующего абонемента (месяц, 3 месяца, 6 месяцев, год);
    - **Month_to_end_contract** — срок до окончания текущего действующего абонемента (в месяцах) *за месяц до проверки факта оттока*;
    - **Group_visits** — факт посещения групповых занятий *за всё время*;
    - **Avg_class_frequency_total** — средняя частота посещений в неделю за все время с начала действия абонемента;
    - **Avg_class_frequency_current_month** — средняя частота посещений в неделю за предыдущий месяц;
    - **Avg_additional_charges_total** — суммарная выручка от других услуг фитнес-центра: кафе, спорт-товары, косметический и массажный салон.

### 2.2. Средние значения признаков в двух группах по целевой переменной

```python
# <сводная таблица средним значениям признаков>
avg_by_churn = df.groupby('churn').mean().T
avg_by_churn['difference_1_0'] = avg_by_churn[1] - avg_by_churn[0]
avg_by_churn
```

**gender**, **near_location**, **partner**, **promo_friends**, **phone**, **group_visits** - признаки, принимающие значения 0 и 1. Чем ближе к 1, тем больше единиц в значениях признака группы.

Для этих признаков *mean* соответствует доле единиц относительно общего числа клиентов в выборке.

Видим, что в целом клиенты, не ушедшие в отток (churn = 0) имеют более высокие показатели по большинству признаков:

- Доля **near_location = 1** больше на **10%** — более часто живут или работают рядом с фитнес-центром;
- Доля **partner = 1** больше почти на **18%** — более часто работают в компании-партнере;
- Доля **promo_friends = 1** больше почти на **17%** — их больше присоединилось по промо-коду от знакомых;
- Доля **group_visits = 1** больше на **19.5%** — более часто являются участниками групповых занятий;
- **contract_period**, **month_to_end_contract**, **lifetime** выше — более длительный абонемент, более долгая история сотрудничества;
- **avg_class_frequency_current_month**, **avg_class_frequency_total** выше — чаще посещают фитнес-центр;
- **avg_additional_charges_total** выше  — совершают больше дополнительных покупок.

Признаки **gender**, **age** и **phone** более сложны для интерпретации.

- **gender** распределен практически одинаково в двух группах;
- **age**: клиенты, не ушедшие в отток, в среднем *на 3 года старше*;
- **phone** распределен практически одинаково в двух группах.

## 3. Модель прогнозирования оттока клиентов

### 3.1. Разбиение на обучающую и валидационную выборки

```python
# <разделение данных на признаки (матрица X) и целевую переменную (y)>
X = df.drop('churn', axis = 1)
y = df['churn']
```

```python
# <разделение модели на обучающую и валидационную выборку>
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# <проверка размера выборок>
print('Обучающая выборка:', len(X_train))
print('Валидационная выборка:', len(X_test))
print('Сравнение обучающей выборки с размером всего датасета:', len(X_train)/len(X))
```

### 3.2. Обучение модели логистической регрессией и случайным лесом

```python
# <модель логистической регрессии>
model_l1 = LogisticRegression(random_state=0, solver='liblinear')
model_l1.fit(X_train, y_train)
predictions_l1 = model_l1.predict(X_test)
probabilities_l1 = model_l1.predict_proba(X_test)[:,1]
```

```python
# <модель случайного леса, количество деревьев - 50>
model_rf1 = RandomForestClassifier(n_estimators=50, random_state=0)
model_rf1.fit(X_train, y_train)
predictions_rf1 = model_rf1.predict(X_test)
probabilities_rf1 = model_rf1.predict_proba(X_test)[:,1]
```

```python
# <функция вывода метрик>
def print_all_metrics(y_true, y_pred, y_proba, name):
    return pd.DataFrame([accuracy_score(y_true, y_pred), precision_score(y_true, y_pred), recall_score(y_true, y_pred),
                        f1_score(y_true, y_pred), roc_auc_score(y_true, y_proba)], 
                        index=['accuracy', 'precision', 'recall', 'F1', 'ROC_AUC'], columns=[name])
```

```python
# <метрики для логистической регрессии>
metrics = print_all_metrics(y_test, predictions_l1, probabilities_l1, name='ЛР (все признаки)')
metrics
```

```python
# <метрики для случайного леса>
metrics=metrics.join(print_all_metrics(y_test, predictions_rf1, probabilities_rf1, name='СЛ (все признаки, 50)'))
metrics
```

В базовом случае логистическая регрессия показала себя чуть лучше, чем случайный лес.

Попробуем увеличить число деревьев для алгоритма случайного леса.

```python
# <модель случайного леса, количество деревьев - 100>
model_rf2 = RandomForestClassifier(n_estimators=100, random_state=0)
model_rf2.fit(X_train, y_train)
predictions_rf2 = model_rf2.predict(X_test)
probabilities_rf2 = model_rf2.predict_proba(X_test)[:,1]
```

```python
# <метрики для случайного леса>
metrics=metrics.join(print_all_metrics(y_test, predictions_rf2, probabilities_rf2, name='СЛ (все признаки, 100)'))
metrics
```

В результате изменения количества деревьев метрики стали даже хуже.

## 5. Выводы и рекомендации по работе с клиентами

**Общие выводы**

Чем больше клиент посещает фитнес-центр и чем активней он участвует в акциях и иных активностях, тем более вероятно, что он не уйдет в отток.

Наиболее надежные клиенты:
- Приходят по приглашению знакомых и, по-видимому, совместно посещают фитнес-центр;
- Являются сотрудниками компаний-партнеров;
- Активно посещают групповые занятия;
- Оформляют длительные абонементы;
- Часто посещают фитнес-центр.

Клиенты, наиболее часто уходящие в отток, обычно не участвуют в этих активностях.

Выяснилось, что проживание или работа рядом с фитнес-центром не всегда приводит к высокой надежности клиента. Об этом свидетельствует кластер 3. Все клиенты из этого кластера проживают или работают рядом с фитнес-центром, однако редко посещают фитнес-центр и часто уходят в отток.

В то же время если клиент проживает или работает не в районе фитнес центра (кластер 2), по умолчанию велика вероятность того, что он уйдет в отток.

**Рекомендации для стратегии взаимодействия с клиентами и их удержания:**

То, что уже делается:

- Регулярное проведение акции «Приведи друга». Оформление первого абонемента по ссылке знакомого повышает вероятность долгого сотрудничества. Во-первых, новый клиент приходит по рекомендации знакомого и больше доверяет предоставляемым услугам. Во-вторых, у нового клиента есть возможность заниматься вместе с друзьями.


- Добавление разнообразия типов групповых занятий. Включить занятия по разным видам йоги, кроссфиту и другим модным активностям, которых еще нет в расписании фитнес-центра. Клиенты, посещающие групповые занятия, реже уходят в отток.

Дополнительно:

- Поощрение оформления длительных абонементов (*6 и 12 месяцев*). Возможно предоставление скидки и/или рассрочки. Клиенты, оформившие длительные абонементы, склонны к долгосрочному сотрудничеству.


- Разработка и внедрение системы поощрения регулярных посещений фитнес-центра. Возможно небольшое вознаграждение за выполнение плана тренировок на месяц либо разработка фитнес-приложения для клиентов по отслеживанию спортивных достижений за период. Регулярно посещающий фитнес-центр клиент с меньшей вероятностью уйдет в отток.
