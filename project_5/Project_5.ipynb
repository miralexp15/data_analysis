```python
# <замена типов данных>
visits_log['end_ts'] = pd.to_datetime(visits_log['end_ts'])
visits_log['start_ts'] = pd.to_datetime(visits_log['start_ts'])
# <вывод первых пяти строк>
visits_log.head()
```

## 2. Метрики

### 2.1. Продуктовые метрики

#### 2.1.1. Оценка пользовательской активности

Необходимо оценить количество визитов в день (*DAU*), в неделю (*WAU*) и в месяц (*MAU*). Данные возьмем из таблицы *visits_log*.

```python
# <первые 5 строк таблицы>
visits_log.head()
```

Добавим в таблицу *visits_log* дополнительные колонки: год (*session_year*), месяц (*session_month*), неделя (*session_week*), дата (*session_date*).

```python
# <добавление года, месяца, недели и даты>
visits_log['session_year']  = visits_log['start_ts'].dt.year
visits_log['session_month'] = visits_log['start_ts'].dt.month
visits_log['session_week']  = visits_log['start_ts'].dt.week
visits_log['session_date'] = visits_log['start_ts'].dt.date
visits_log.head()
```

Считаем DAU, WAU и MAU.

```python
# <daily active users>
dau = visits_log.groupby('session_date').agg({'uid':'nunique'})
dau.head()
```

```python
# <weekly active users>
wau = visits_log.groupby(['session_year', 'session_week']).agg({'uid':'nunique'})
wau.head()
```

```python
# <monthly active users>
mau = visits_log.groupby(['session_year', 'session_month']).agg({'uid':'nunique'})
mau.head()
```

Найдем средние значения метрик.

```python
print('Среднее количество уникальных пользователей в день (DAU): {:.2f}'.format(float(dau.mean())))
print('Среднее количество уникальных пользователей в неделю (WAU): {:.2f}'.format(float(wau.mean())))
print('Среднее количество уникальных пользователей в месяц (MAU): {:.2f}'.format(float(mau.mean())))
```

Изменение метрик во времени.

```python
# <DAU>
plt.style.use('ggplot') # <настройка стиля графиков>
ax = dau['uid'].plot(figsize=(15, 10), style='.-', c='blue', mfc='red', mec='red', rot=45)
plt.title('DAU, количество уникальных пользователей в день')
plt.xlabel('Дата сессий')
plt.ylabel('Количество уникальных посетителей')
plt.yticks(np.arange(0, 3401, 200))
plt.xticks([datetime(2017,5,1) + relativedelta(months=i) for i in range(15)],
           ['']+[(datetime(2017,6,1) + relativedelta(months=i)).strftime('%Y-%m-%d') for i in range(13)]+['']);
```

```python
# <ТОП-2 аномально высоких значений>
dau['uid'].sort_values(ascending=False).head(2)
```

Наиболее высокое значение посетителей зафиксировано *24 ноября 2017 года* - **3319** посетителей (на *1332* больше второго дня по количеству уникальных посетителей).

```python
# <дни с аномально низким количеством посетителей>
dau[dau['uid'] < 200]
```

По данному графику делаем следующие выводы:

#### 2.1.2. Количество заходов на сайт в день

```python
# <первые пять строк датафрейма с визитами>
visits_log.head()
```

Чтобы найти среднее количество заходов пользователей на сайт в день, нужно сгруппировать данные по дате и идентификатору пользователя и посчитать количество заходов. После этого найти среднее значение для каждой даты.

Будем считать эту метрику только для тех пользователей, которые в конкретный день заходили на сайт хотя бы раз.

```python
# <количество заходов в день>
visits_per_day = visits_log.groupby(['session_date', 'uid']).agg({'uid':'count'})
visits_per_day.columns = ['visits_per_user']
visits_per_day.reset_index(inplace=True)
visits_per_day.head()
```

```python
# <среднее количество посещений в день>
visits_per_day_mean = visits_per_day.groupby('session_date')['visits_per_user'].mean().reset_index()
visits_per_day_mean.head()
```

```python
# <общая характеристика столбца с визитами>
visits_per_day_mean['visits_per_user'].describe()
```

В среднем пользователи заходят около *одного раза в день*.

#### 2.1.4. Расчет Retention Rate

Применим когортный анализ. Данные охватывают один год. Сгруппируем пользователей по 12 когортам. Событие для группировки: первая покупка, период: месяц.

Проведем расчет *Retention Rate* для таблице с заказами (*orders_log*).

```python
# <вывод первых пяти строк>
orders_log.head()
```

```python
# <создадим таблицу с месяцем первой покупки для каждого пользователя>
first_order_date = orders_log.groupby('uid')['buy_ts'].min().dt.date.apply(lambda x: x.strftime('%Y-%m')).reset_index()
first_order_date.columns = ['uid', 'first_order_month']
first_order_date.head()
```

```python
# <объединение таблиц>
orders_with_cohorts = orders_log.merge(first_order_date, on='uid', how='left')
orders_with_cohorts.sample(5, random_state=11)
```

```python
# <общая оценка показателей когорт>
orders_with_cohorts.groupby('first_order_month').agg({'uid':'nunique'})
```

В июне 2018 года были сделан лишь один заказ. Объединим эту когорту с предыдущей.

```python
# <объединение когорт>
orders_with_cohorts.loc[orders_with_cohorts['first_order_month'] == '2018-06', 'first_order_month'] = '2018-05'
orders_with_cohorts.loc[orders_with_cohorts['buy_ts'] == '2018-06-01 00:02:00', 'buy_ts'] = datetime(2018, 5, 31, 23, 59)
orders_with_cohorts.tail(1)
```

Теперь добавим в таблицу столбец *cohort_lifetime*.

```python
# <добавление столбца 'cohort_lifetime' в таблицу>
orders_with_cohorts['activity_month'] = orders_with_cohorts['buy_ts'].apply(lambda x: x.strftime('%Y-%m'))
temp_next = orders_with_cohorts['activity_month'].apply(lambda x: datetime.strptime(x, '%Y-%m'))
temp_first = orders_with_cohorts['first_order_month'].apply(lambda x: datetime.strptime(x, '%Y-%m'))
orders_with_cohorts['cohort_lifetime'] = (temp_next.dt.to_period("M") - temp_first.dt.to_period("M")).map(lambda x: x.n)
orders_with_cohorts.sample(5, random_state=11)
```

Теперь подготовим таблицу **cohorts_counts** c расчетами по когортам.

```python
# <количество пользователей по когортам>
cohorts_counts = orders_with_cohorts.groupby(['first_order_month', 'cohort_lifetime']).agg({'uid':'nunique'}).reset_index()
cohorts_counts.columns = ['first_order_month', 'cohort_lifetime', 'uid_counts']
cohorts_counts.sample(5, random_state=11)
```

Осталось добавить колонку с общим количеством активных пользователей по когортам и *retention*.

```python
# <количество пользователей по когортам>
temp = orders_with_cohorts.groupby('first_order_month').agg({'uid':'nunique'}).reset_index()
temp.columns = ['first_order_month', 'users_total']
temp
```

```python
# <объединение таблиц и подсчет retention>
cohorts_counts = cohorts_counts.merge(temp, how='left', on='first_order_month')
cohorts_counts['retention'] = cohorts_counts['uid_counts'] / cohorts_counts['users_total']
cohorts_counts.sample(5, random_state=11)
```

Теперь создадим сводную таблицу по *Retention Rate*.

```python
# <сводная таблица для Retention Rate>
retention_pivot = cohorts_counts.pivot_table(index='first_order_month', columns='cohort_lifetime', 
                                             values='retention', aggfunc='sum')
retention_pivot
```

Для наглядности составим тепловую карту.

```python
# <тепловая карта удержания клиентов>
sns.set(style='white')
plt.figure(figsize=(13, 9))
sns.heatmap(retention_pivot, annot=True, fmt='.1%', linewidths=1, linecolor='gray')
plt.yticks(rotation=0)
plt.title('Тепловая карта удержания клиентов')
plt.xlabel('Lifetime когорт')
plt.ylabel('Когорта');
```

```python
# <тепловая карта удержания клиентов без 0го периода lifetime>
sns.set(style='white')
plt.figure(figsize=(13, 9))
plt.title('Cohorts: User Retention')
sns.heatmap(retention_pivot.iloc[:, 1:], annot=True, fmt='.1%', linewidths=1, linecolor='gray')
plt.yticks(rotation=0)
plt.title('Тепловая карта удержания клиентов без нулевого периода')
plt.xlabel('Lifetime когорт')
plt.ylabel('Когорта');
```

Наблюдаем, что процент удержания клиентов достаточно невысокий. Уже на lifetime №1 ни для одной когорты результат не превышает *6%*.

Скорее всего это говорит о том, что пользователи нерегулярно посещают культурные мероприятия. Когда проводится особо интересующее людей событие, они покупают билеты, в остальное время остаются неактивны.

Видим, что наилучшее удержание пользователей на большой дистанции показывает первая когорта. Скорее всего в эту когорту попали постоянные клиенты сервиса, поэтому Retention Rate для этой когорты (единственной) не падает ниже *2%*.

Дополнительно отметим, что в каждой последующей когорте Retention Rate падает до значений менее 2% быстрее предыдущей. Таким образом, показатель Retention Rate ухудшается.

```python
# <изменение Retention Rate на протяжении lifetime>
plt.style.use('ggplot') # <настройка стиля графиков>
temp = retention_pivot.T.iloc[1:, :-2]*100
temp.columns.name = 'Когорты'
temp.plot(figsize=(15, 9), linewidth=2, colormap=cm.tab20)
plt.title('График изменений Retention Rate')
plt.xlabel('Lifetime')
plt.ylabel('Retention Rate, %')
plt.xlim(0.5, 11.5)
plt.xticks(np.arange(1,12))
plt.yticks(np.arange(1,6,0.5));
```

Данный график менее удобен для оценки изменения Retention Rate во времени, чем тепловая карта. Однако на нем также отслеживается тренд на ускорение падения Retention Rate для новых когорт.

Расчитаем средний Retention Rate для второго месяца жизни (lifetime = 1).

```python
# <средний Retention Rate для lifetime=1>
retention_pivot.iloc[:-1, 1].mean() * 100
```

Средний Retention Rate для второго месяца жизни когорт составляет около **3.7%**.

```python
# <сумма покупок по пользователям для каждой когорты и месяца>
cohorts_months_revenue_sum = orders_with_cohorts.groupby(['first_order_month', 'cohort_lifetime'])['revenue'].sum().reset_index()
cohorts_months_revenue_sum.head()
```

#### 2.3.3. ROMI

Рассчитаем ROMI (Return On Marketing Investments) по когортам в разрезе источников.

*ROMI равен LTV / CAC*.

Для начала нужны 2 таблицы:
- LTV на когорту по месяцу и источнику;
- CAC на когорту по источнику.

```python
# <вывод первых пяти строк таблицы с заказами и источником>
visits_and_orders.head()
```

```python
# <подготовка таблицы для вычислений>
temp = visits_and_orders.groupby('uid')['buy_ts'].min().dt.strftime('%Y-%m')
temp.name = 'first_order_date'
visits_and_orders = visits_and_orders.join(temp, on='uid', how='left')
visits_and_orders['order_date'] = visits_and_orders['buy_ts'].dt.strftime('%Y-%m')
visits_and_orders.sample(5, random_state=11)
```

```python
# <количество клиентов в когортах по источникам>
count_users_sources_cohorts = visits_and_orders.groupby(['source_id', 'first_order_date'])['uid'].nunique()
count_users_sources_cohorts.head()
```

```python
# <ltv по источникам, когортам и месяцам>
ltv_sources_cohorts_months = visits_and_orders.groupby(['source_id', 'first_order_date', 'order_date']).agg({'revenue':'sum'}).reset_index()
ltv_sources_cohorts_months.set_index(['source_id', 'first_order_date'], inplace=True)
ltv_sources_cohorts_months = ltv_sources_cohorts_months.join(count_users_sources_cohorts, how='left')
ltv_sources_cohorts_months['ltv_cumsum'] = ltv_sources_cohorts_months.groupby(['source_id', 'first_order_date'])['revenue'].cumsum()
ltv_sources_cohorts_months.tail(10)
```

```python
ltv_sources_cohorts_months['ltv'] = ltv_sources_cohorts_months['ltv_cumsum'] / ltv_sources_cohorts_months['uid']
ltv_sources_cohorts_months_for_total = ltv_sources_cohorts_months.copy() # <для общих расчетов>
ltv_sources_cohorts_months = ltv_sources_cohorts_months[['order_date', 'ltv']]
ltv_sources_cohorts_months.tail(10)
```

Подготовим таблицу CAC по когортам и источникам.

```python
# <CAC по когортам и источникам>
cac_sources_cohorts = cac_total_sources[['source_id', 'year_and_month', 'cac']]
cac_sources_cohorts.columns = ['source_id', 'first_order_date', 'cac']
cac_sources_cohorts.set_index(['source_id', 'first_order_date'], inplace=True)
cac_sources_cohorts.head()
```

```python
# <таблица romi>
romi_sources_cohorts_months = ltv_sources_cohorts_months.join(cac_sources_cohorts, how='left').reset_index()
romi_sources_cohorts_months['romi'] = romi_sources_cohorts_months['ltv'] / romi_sources_cohorts_months['cac']
romi_sources_cohorts_months.head()
```

```python
# <добавление lifetime для создания сводной таблицы>
temp_first = romi_sources_cohorts_months['first_order_date'].apply(lambda x: datetime.strptime(x, '%Y-%m'))
temp_next = romi_sources_cohorts_months['order_date'].apply(lambda x: datetime.strptime(x, '%Y-%m'))
romi_sources_cohorts_months['cohort_lifetime'] = (temp_next.dt.to_period("M") - temp_first.dt.to_period("M")).map(lambda x: x.n)
romi_sources_cohorts_months.head()
```

```python
# <ROMI по когортам в разрезе источников>
romi_pivoted = romi_sources_cohorts_months.pivot_table(columns='cohort_lifetime', 
                                                       index=['source_id', 'first_order_date'],
                                                       values='romi',
                                                       aggfunc='sum')
romi_pivoted = romi_pivoted.drop(index=7)
romi_pivoted
```

Для наглядности построим heatmap.

```python
# <тепловая карта ROMI по источникам, когортам и месяцам>
for i, j in [[1, 'ax1'], [2, 'ax2'], [3, 'ax3'], [4, 'ax4'], [5, 'ax5'], [9, 'ax9'], [10, 'ax10']]:
    plt.figure(figsize=(13, 9))
    globals()[j] = sns.heatmap(romi_pivoted.loc[i], annot=True, fmt='.1%', linewidths=1, linecolor='gray')
    plt.yticks(rotation=0)
    plt.title('Тепловая карта ROMI по когортам и месяцам для источника № ' + str(i))
    plt.xlabel('Lifetime')
    plt.ylabel('Когорта')
```

Такие подробные тепловые карты пригодятся для более подробного анализа.

В целом ROMI отличается от месяца к месяцу и по источникам.

В среднем наиболее успешные когорты относятся к когортам начала исследуемого периода, так как в тех когортах присутствуют наиболее лояльные (постоянные) клиенты.

- Пользовательская активность нелинейно изменялась в течение года. Рост с малых значений (начало роста - *сентябрь 2017*) до пика в *ноябре 2017* и дальнейший плавный спад. Рост активности ожидается в *сентябре 2018*;
- В среднем пользователи заходят один раз в день. Скорее всего они заходят исключительно с целью купить билеты. Дополнительных факторов (журнал, новости, онлайн-мероприятия), удерживающих посетителя на сайте нет;
- Типичная пользовательская сессия имеет длительность 5 минут;
- Процент удержания клиентов достаточно невысокий. Пользователи совершают покупку и в дальнейшем нечасто пользуются предлагаемыми услугами. Самая перспективная когорта - **2017-06**, так как в нее входит наибольшее количество постоянных клиентов. В каждой последующей когорте Retention Rate падает до значений менее 2% быстрее предыдущей.

2. Метрики электронной коммерции

- В среднем с момента первого посещения сайта до совершения покупки проходит **16 минут**, т.е. большая часть клиентов в первый раз приходят на сайт с целью приобрести билет на мероприятие. Однако есть клиенты, которые ожидают либо требуемого мероприятия, либо выгодных предложений;
- Около **79.5%** клиентов сделали всего **одну покупку за 6 месяцев**. Максимальное количество покупок - *160*. *Среднее количество покупок на одного покупателя* составляет около **1.44 покупок**. В основном клиенты заинтересованы в билетах на одно конкретное мероприятие;
- Средний чек за всё время составляет около 5 у.е.;
- *Наиболее успешными по LTV* являются когорты **2017-06** и **2017-09**. В когорту **2017-06** попали в том числе *постоянные клиенты*, чем и обусловлены хорошие результаты. В когорты **2017-09** заметен всплеск активности на *lifetime = 3*. За счет него когорта и показала выделяющиеся результаты. *Средний LTV за шесть месяцев* составляет **8 у.е.**

3. Маркетинговые метрики
- *Основная доля расходов на маркетинг* приходится на **третий источник**. В то же время *по показателю ROMI* он показал *наихудший результат*. Следует изучить, в чем причина таких результатов для данного источника трафика;
- График расходов на маркетинг напоминает график пользовательской активности (*MAU*), что скорее всего является совпадением. Такую форму графика определяет наиболее затратный и наименее эффективный источник трафика - **источник 3**;
- Средний **САС** составляет около **9 у.е.**, что больше **LTV** на **1 у.е.**. *Самые эффективные по стоимости привлечения одного клиента*: источники **10**, **9** и **4**, *наименее эффективные*: источники **2** и **3**;
- Источник трафика, не отбивший расходы ни для одной когорты: **источник 3**. 
- По общему ROMI неплохо себя показывают две когорты: **2017-06** и **2017-09**. Лишь для этих двух когорт достигается окупаемость инвестиций. *Для остальных когорт окупаемость ниже допустимого уровня* (ниже 100%).

### 3.3. Итоги когортного анализа

Последние когорты перестали приносить достаточно выручки, чтобы отбивать расходы на маркетинг.

- улучшить информационную, мультимедийную и/или интерактивную составляющую сервиса;
- находить популярные мероприятия, которые могут заинтересовать целевую когорту. Клиенты обычно нацеливаются на конкретное мероприятие;
- изменить приоритетные источники трафика (см. пункт 3.1).
