### 2.2. Столбец last_price (цена на момент снятия с публикации)

```python
# <количество уникальных значений>
df['last_price'].nunique() 
```

```python
# <общее описание значений столбца 'last_price'>
df.last_price.describe() 
```

Минимальное значение составляет **12 190 рублей**, что является аномальным значением и требует проверки.<br><br>
Максимальное значение составляет **763 млн рублей**, что может быть корректным значением.

Преобразуем значения в целочисленный тип.

```python
# <преобразование float в int>
df['last_price'] = df.last_price.astype('int')
# <так как максимальная сумма равна 763 млн, допустимо использовать тип данных int32>
df.last_price.dtype 
```

Проверим, какое количество квартир продавалось менее, чем за а) 1 млн рублей, б) 100 тыс рублей.

```python
# <количество квартир, проданных за менее чем 1 млн руб>
len(df.query('last_price < 10**6')) 
```

```python
# <количество квартир, проданных за менее чем 100 тыс руб>
len(df.query('last_price < 10**5')) 
```

```python
# <квартира, проданная за менее чем 100 тыс руб>
df.query('last_price < 10**5')
```

Таким образом, цена на момент снятия публикации лишь в одном случае меньше 100 тыс рублей (**12 190**) и в 224 случаях меньше 1 млн (также очень низкая цена для Санкт-Петербурга).<br><br>
Возможные причины появления таких значений:
- Такая сумма была выставлена по ошибке;
- Малая сумма была указана для привлечения внимания потенциальных покупателей;
- Мошеннические операции с недвижимостью;

Создадим функцию для определения категории этажа квартиры.

```python
# <создание столбца 'floor_cat' с применением функции 'floor_cat_func'>
df['floor_cat'] = df.apply(floor_cat_func, axis=1) 
# <проверка значений столбца>
df['floor_cat'].value_counts(dropna=False) 
```

#### 4.3.4. Зависимость цены от этажа квартиры

Необходимо проанализировать связь столбцов **last_price**, **price_per_sqm** и **floor_cat**.<br><br>
Изучим, зависит ли общая цена квартиры и цена за квадратный метр от этажа квартиры (Первый, Последний, Другой).

```python
# <общие характеристики для категорий 'floor_cat'>
temp_list = ['Первый', 'Последний', 'Другой']
for i in temp_list:
    print('Характеристики для этажа "{}":\n {}\n'
         .format(i, df.query('floor_cat == @i')[['last_price', 'price_per_sqm']]
                 .apply([min, max,'mean', 'median']).astype(int)))
```

```python
# <floor_cat_ch - собранные вместе общие характеристики>
floor_cat_ch = df.groupby('floor_cat')[['last_price', 'price_per_sqm']].agg([min, max,'mean', 
                                                                             'median']).astype(int).loc[['Первый', 
                                                                                                         'Последний', 
                                                                                                         'Другой']]
floor_cat_ch.columns = floor_cat_ch.columns.to_flat_index() # <удаление мультииндекса>
floor_cat_ch = floor_cat_ch.transpose() # <транспонируем таблицу>
floor_cat_ch
```

```python
# <построение горизонтальных столбчатых диаграмм>
fig, ((ax0, ax1), (ax2, ax3), (ax4, ax5), (ax6, ax7)) = plt.subplots(4, 2, figsize=(15,10))
plt.subplots_adjust(left=0.1, bottom=1, right=0.9, top=2.5, wspace=0.4, hspace=0.4)
plt.suptitle('Значения статистик цены и цены за квадратный метр для разных этажей', y=2.55, fontsize=20)


for i in range(8):
    temp = globals()['ax' + str(i)]
    temp.barh(floor_cat_ch.columns, floor_cat_ch.iloc[i], color='#2F4F4F')
    temp.grid()
    temp.set_xlabel(floor_cat_ch.index[i])
    temp_xticks = np.sort(list(temp.get_xticks()) + list(floor_cat_ch.iloc[i]))
    temp_xticks = [int(temp_xticks[i]) for i in range(1, len(temp_xticks)) if temp_xticks[i] -
                   temp_xticks[i-1] > 0.1*(temp_xticks[1])]
    temp.set_xticks(temp_xticks)
    temp.set_xticklabels(temp_xticks, rotation=90)
    #temp.xticks(rotation=45)
```

Выводы на основе таблицы и графика:
- У объектов недвижимости **на первом этаже** самая низкая медиана и среднее;
- У объектов недвижимости **на последнем этаже** самая высокая медиана цены; самое высокое среднее и медиана цены за квадратный метр;
- У объектов недвижимости **на других этажах** самое высокое среднее цены. Это связано с тем, что максимальная цена этой группы составляет **763 млн рублей**, что является аномально высоким значением и смещает среднее группы.

Теперь посчитаем количество значений, выпадающих из промежутка внутри "усов" boxplot (Q1 - 1.5 МР, Q3 + 1.5 МР) для каждой группы, а ткаже долю выпадающих значений.

```python
# <функция для определения "усов" boxplot>
def whisker(df, col):
    if (df[col].quantile(0.25) - 1.5 * (df[col].quantile(0.75) - df[col].quantile(0.25))) < df[col].min():
        return [*(df[col].min(), df[col].quantile(0.75) + 1.5 * (df[col].quantile(0.75) - df[col].quantile(0.25)))]
    else:
        return [*(df[col].quantile(0.25) - 1.5 * (df[col].quantile(0.75) - df[col].quantile(0.25)),
                df[col].quantile(0.75) + 1.5 * (df[col].quantile(0.75) - df[col].quantile(0.25)))]
```

```python
temp_list = ['Первый', 'Последний', 'Другой']

# <1 last_price>
print('Цена')
for i in temp_list:
    temp_df = df.query('~last_price.isna() and floor_cat == @i')
    temp = whisker(temp_df, 'last_price')
    temp_len = (len(temp_df[temp_df.last_price < temp[0]]), len(temp_df[temp_df.last_price > temp[1]]))
    print('{} -'.format(i), temp_len,'\nДоля выпадающих:', (temp_len[0]+temp_len[1])/len(temp_df) )
    
# <2 price_per_sqm>
print('\nЦена за квадратный метр')
for i in temp_list:
    temp_df = df.query('~last_price.isna() and floor_cat == @i')
    temp = whisker(temp_df, 'price_per_sqm')
    temp_len = (len(temp_df[temp_df.price_per_sqm < temp[0]]), len(temp_df[temp_df.price_per_sqm > temp[1]]))
    print('{} -'.format(i), temp_len,'\nДоля выпадающих:', (temp_len[0]+temp_len[1])/len(temp_df) )
```

### 4.4. Цена объектов в разных населенных пунктах

Необходимо проанализировать связь столбцов **last_price**, **price_per_sqm** и **locality_name**.<br><br>
Для начала исследуем значения столбца **locality_name**.

```python
# <количество уникальных значений>
df.query('~locality_name.isna()').locality_name.nunique() 
```

```python
# <определение уникальных типов населенных пунктов>
def unique_area_types():
    name_types = []
    for words in df.query('~locality_name.isna()').locality_name.unique():
        temp_str = ''
        for word in words.split(' '):
            if word != word.title():
                temp_str += word + ' '
        temp_str = temp_str.rstrip()
        name_types.append(temp_str)
    return np.sort(pd.Series(name_types).unique())
```

```python
# <уникальные типы населенных пунктов>
unique_area_types()
```

Обращаем внимание, что присутствуют 2 варианта одного слова: **посЕлок** и **посЁлок**. Сделаем написание единообразным, заменив все буквы **Ё** на **Е**.

```python
# <замена Ё на Е>
df['locality_name'] = df.loc[~df['locality_name'].isna(), 'locality_name'].apply(lambda x: x.replace('ё', 'е'))
unique_area_types()
```

```python
# <количество уникальных значений уменьшилось на 34>
df.query('~locality_name.isna()').locality_name.nunique() 
```

```python
# <проверка соответствия названия населенного пункта и его типа>
names = []
for words in df.query('~locality_name.isna()').locality_name.unique():
    temp_str = ''
    for word in words.split(' '):
        if word == word.title():
            temp_str += word + ' '
    temp_str = temp_str.rstrip()
    names.append([temp_str, words])
names = pd.DataFrame(names, columns=['names', 'full_names'])

names_count = names.groupby('names')['full_names'].count().sort_values(ascending=False)

names_count = names_count.where(lambda x: x > 1).dropna().astype(int)
print('\nКоличество совпадений названий -', len(names_count))
names_count
```

```python
# <таблица совпадающих названий>
sub_names = names[names['names'].isin(list(names_count.index))].sort_values(by='names')
sub_names.head(7)
```

Теперь выберем значения для замены. Для этого объединим столбец **full_names** таблицы **names** с помощью **groupby** и метода **first**, а затем создадим общую таблицу замены **fill_sub_names**.

```python
# <значения для замены>
sub_temp = sub_names.groupby('names')['full_names'].first()
sub_temp.name = 'sub'
fill_sub_names = sub_names.merge(sub_temp, how='left', on='names')
fill_sub_names.head()
```

```python
# <функция для замены>
def fill_names(x):
    if fill_sub_names['full_names'].str.contains(x).any():
        return fill_sub_names.loc[fill_sub_names['full_names'] == x]['sub'].values[0]
    else:
        return x
```

```python
# <замена типов населенных пунктов>
df['locality_name'] = df.loc[~df['locality_name'].isna(), 'locality_name'].apply(fill_names)
df.query('~locality_name.isna()').locality_name.nunique() # <количество уникальных значений>
```

*Построение столбчатых диаграмм для квантилей распределения*

```python
# <подготовка данных>
lp_first = df_center.query('floor_cat == "Первый" and ~last_price.isna()')['last_price']
lp_last = df_center.query('floor_cat == "Последний" and ~last_price.isna()')['last_price']
lp_other = df_center.query('floor_cat == "Другой" and ~last_price.isna()')['last_price']
```

```python
# <1 для цены>
lp_first_concat = pd.concat([lp_first, pd.qcut(lp_first, 10)], axis=1)
lp_first_concat.columns =['Первый', 'quantile']
lp_first_concat_mean = lp_first_concat.groupby('quantile').agg(['mean', 'median']).set_index(np.arange(1, 11, 1))

lp_last_concat = pd.concat([lp_last, pd.qcut(lp_last, 10)], axis=1)
lp_last_concat.columns =['Последний', 'quantile']
lp_last_concat_mean = lp_last_concat.groupby('quantile').agg(['mean', 'median']).set_index(np.arange(1, 11, 1))

lp_other_concat = pd.concat([lp_other, pd.qcut(lp_other, 10)], axis=1)
lp_other_concat.columns =['Другой', 'quantile']
lp_other_concat_mean = lp_other_concat.groupby('quantile').agg(['mean', 'median']).set_index(np.arange(1, 11, 1))

quantiles_floors_lp = lp_first_concat_mean.join([lp_last_concat_mean, lp_other_concat_mean])
quantiles_floors_lp
```
