#### 2.2.3. Столбец year_of_release

В данном столбце необходимо обработать пропущенные значения и поменять тип данных с *float* на *int*.

```python
# <количество охваченных лет>
df['year_of_release'].nunique() 
```

```python
# <количество пропущенных значений>
df['year_of_release'].isna().sum() 
```

```python
# <количество пропусков для разных платформ>
platforms_with_na = df[df['year_of_release'].isna()]['platform'].value_counts()
platforms_with_na
```

```python
# <таблица с размахом годов выхода игр для разных платформ>
platform_years['difference_of_years'] = platform_years['last'] - platform_years['first']
platform_years.head()
```

Рассмотрим только строки с актуальными платформами (выход последней игры на платформе - 2016 год). Менее важные для нас значения не будем заполнять, чтобы избежать смещения распределения выхода игр по годам. Для определения важных платформ используем таблицу **platform_years**.

```python
# <создание таблицы пропущенных значений на актуальных платформах>
platform_years_selected = platform_years[platform_years['last'] == 2016].join(platforms_with_na)
platform_years_selected.columns = ['first', 'last', 'difference_of_years', 'missed_count']
platform_years_selected.sort_values(by='missed_count')
```

```python
# <количество пропущенных значений для актуальных платформ>
platform_years_selected.missed_count.sum() 
```

Видим, что большАя часть пропусков относится к интересующим нас в первую очередь платформам.

**tbd** в столбце **user_score** переводится как *To Be Determined* (ссылка: https://www.metacritic.com/faq#item13).<br><br>
Чтобы обеспечить точность оценок для игр, есть следующая система: если количество оценок меньше порога (для сайта **metacritic.com** *4*), то рейтинг не отображается. Вместо рейтинга игры отображается строка **tbd**. Оценка будет отображена только по достижении минимального количества проголосовавших. 

Ячейки с *tbd* какими-либо усредненными значениями заполнять не будем. В конце данного этапа заменим эти значения на **-1**, чтобы выделить их в отдельную группу и преобразовать значения столбца **user_score** к типу *float*.

Проверим количество пропусков в столбцах **critic_score** и **user_score** одновременно.

```python
# <количество одновременных пропусков или 'tbd' в столбцах 'critic_score' и 'user_score'>
((df['critic_score'].isna()) & ((df['user_score'].isna()) | (df['user_score'] == 'tbd'))).sum()
```

```python
# <количество одновременных пропусков в столбцах 'critic_score' и 'user_score'>
((df['critic_score'].isna()) & (df['user_score'].isna())).sum()
```

```python
# <количество пропусков или 'tbd' в одном из двух столбцов с рейтингом>
np.logical_xor(df['critic_score'].isna(), (df['user_score'].isna()) | (df['user_score'] == 'tbd')).sum()
```

```python
# <количество пропусков в одном из двух столбцов с рейтингом>
np.logical_xor((df['critic_score'].isna()) & (df['user_score'] != 'tbd'), (df['user_score'].isna())).sum()
```

Таким образом, для **8 004** игр нет никаких оценок.

**1693** пропуска или *tbd* есть только в одном из двух столбцов.

**611** пропусков, возможно, удастся заполнить с использованием линии регрессии. Для проверки данного способа исследуем, есть ли корреляция между значениями оценок критиков и пользователей.

```python
# <проверка коэффициента корреляции столбцов 'critic_score' и 'user_score'>
with pd.option_context('mode.chained_assignment', None):
    temp_df = df[(~df['critic_score'].isna()) & ((~df['user_score'].isna()) & (df['user_score'] != 'tbd'))]
    temp_df.loc[:, 'user_score'] = pd.to_numeric(temp_df.loc[:, 'user_score'])
    print(temp_df['user_score'].dtype)
temp_df['critic_score'].corr(temp_df['user_score'])
```

Коэффициент корреляции положительный и довольно высокий.

Проведем расчет линии регрессии.

```python
# <построение диаграммы рассеяния и линии регрессии>
temp_df.plot(x='critic_score', y='user_score', kind='scatter', alpha=0.15, figsize=(9, 9))
x = temp_df['critic_score']
y = temp_df['user_score']
m, b = np.polyfit(x, y, 1) # <подсчет коэффициентов линии регрессии>
plt.plot(x, m*x+b, color='orange')
plt.title('Диаграмма рассеяния и линия регрессии для оценок критиков и пользователей');
plt.xlabel('Оценка критиков (critic_score)')
plt.ylabel('Оценка пользователей (user_score)');
```

На графике рассеяния видим, что часто критики и пользователи обычно сходятся во мнении, однако разброс значений довольно велик. Поэтому не будем заполнять пропуски, чтобы объективно проанализировать имеющиеся данные.

```python
# <5 первых строк общей таблицы>
df_act.head()
```

```python
# <таблица для исследования>
df_act_hyp1 = df_act.query('(platform == "XOne" or platform == "PC") and ~user_score.isna() and user_score != -1')
df_act_hyp1.head()
```

```python
# <расчет статистик для платформ PC и Xbox One>
df_act_hyp1.groupby('platform')['user_score'].agg(['min', 'max', 'mean', 'median', 'var', 'std'])
```

Наложим одну гистограмму на другую, чтобы сравнить их по доле значений. Пересчет доли значений - в параметре **weights**.

```python
# <сравнение распределений рейтингов для платформ PC и XboxOne>
plt.subplots(1, 1, figsize=(15,10))
plt.title('Сравнение распределений оценок игроков для платформ PC и XboxOne')
for i, j, k in [['PC', 'tab:blue', 30], ['XOne', 'tab:orange', 30]]:
    temp = df_act_hyp1.loc[df_act_hyp1['platform'] == i, 'user_score']
    temp.hist(bins=k, alpha=0.4, color=j, weights=np.ones_like(np.arange(len(temp))) / len(temp))
plt.legend(['PC', 'Xbox One'], fontsize=20)
plt.xlabel('Оценки')
plt.ylabel('Доля значений')
```

Распределения похожи на нормальные с длинным левым хвостом и коротким правым.

*В итоге:*
- **Формулируем нулевую гипотезу о равенстве средних двух генеральных совокупностей**;
- **Формулируем двустороннюю альтернативную гипотезу**, так как среднее значение для платформы **XOne** ненамного больше средней оценки для **PC**;
- Значение **требуемого уровня значимости принимаем равным 0.05**.

**Нулевая гипотеза (H0)**: средняя оценка для игр на платформе **Xbox One** *равна* средней оценке для игр на платформе **PC**;

**Альтернативная гипотеза (Н1)**: средняя оценка для игр на платформе **Xbox One** *не равна* средней оценке для игр на платформе **PC**.

```python
# <подготовка таблиц для статистического теста>
df_xone = df_act_hyp1.query('platform == "XOne"')['user_score']
df_pc = df_act_hyp1.query('platform == "PC"')['user_score']
```

```python
# критический уровень статистической значимости
alpha = 0.05 
```

```python
# <проверка гипотезы о равенстве средних двух независимых совокупностей>
# <дисперсии отличаются в 3 раза, поэтому 'equal_var'=False>
results = st.ttest_ind(df_xone, df_pc, equal_var=False)
print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**P-value** равен **0.155**. Это означает, что:
1. С вероятностью около **15.5%** имеющиеся различия можно получить случайно;
2. P-value (**0.155**) больше значения альфа (**0.05**), поэтому не получается отклонить нулевую гипотезу.

**Средняя оценка для игр на платформе Xbox One равна средней оценке для игр на платформе PC**.

### 5.2. Гипотеза: средние пользовательские рейтинги жанров Action и Sports разные

```python
# <5 первых строк общей таблицы>
df_act.head()
```

```python
# <таблица для исследования>
df_act_hyp2 = df_act.query('(genre == "Action" or genre == "Sports") and ~user_score.isna() and user_score != -1')
df_act_hyp2.head()
```

```python
# <расчет статистик для платформ PC и Xbox One>
df_act_hyp2.groupby('genre')['user_score'].agg(['min', 'max', 'mean', 'median', 'var', 'std'])
```

Наложим одну гистограмму на другую, чтобы сравнить их по доле значений. Пересчет доли значений - в параметре **weights**.

```python
# <сравнение распределений рейтингов для платформ PC и XboxOne>
plt.subplots(1, 1, figsize=(15,10))
plt.title('Сравнение распределений оценок игроков для жанров Action и Sports')
for i, j, k in [['Action', 'tab:blue', 30], ['Sports', 'tab:orange', 30]]:
    temp = df_act_hyp2.loc[df_act_hyp2['genre'] == i, 'user_score']
    temp.hist(bins=k, alpha=0.4, color=j, weights=np.ones_like(np.arange(len(temp))) / len(temp))
plt.legend(['Action', 'Sports'], fontsize=20)
plt.xlabel('Оценки')
plt.ylabel('Доля значений')
```

Распределения похожи на нормальные с длинным левым хвостом и коротким правым. Распределение для жанра **Sports** более широкое и смещено влево относительно распределения жанра **Action**.

*В итоге:*
- **Формулируем нулевую гипотезу о равенстве средних двух генеральных совокупностей**;
- **Формулируем одностороннюю альтернативную гипотезу**, так как среднее значение для жанра **Action** намного больше средней оценки для **Sports**;
- Значение **требуемого уровня значимости принимаем равным 0.05**.

**Нулевая гипотеза (H0)**: средняя оценка для игр жанра **Action** *равна* средней оценке для игр жанра **Sports**;

**Альтернативная гипотеза (Н1)**: средняя оценка для игр жанра **Action** *выше* средней оценки для игр жанра **Sports**.

```python
# <подготовка таблиц для статистического теста>
df_action = df_act_hyp2.query('genre == "Action"')['user_score']
df_sports = df_act_hyp2.query('genre == "Sports"')['user_score']
```

```python
# критический уровень статистической значимости
alpha = 0.05 
```

```python
# <проверка односторонней гипотезы о равенстве средних двух независимых совокупностей>
# <дисперсии сильно отличаются, поэтому 'equal_var'=False>
results = st.ttest_ind(df_action, df_sports, equal_var=False)
print('p-значение:', results.pvalue / 2)
if results.pvalue / 2 < alpha and df_action.mean() > df_sports.mean():
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**P-value** примерно равен **1**. Это означает, что:
1. С вероятностью около **0%** имеющиеся различия можно получить случайно;
2. P-value ($p=6.94\cdot10^{-21}$) меньше значения альфа (**0.05**), поэтому отклоняем нулевую гипотезу.

**Средняя оценка для игр жанра Action (ГС1) выше средней оценки для игр жанра Sports (ГС2)**.
