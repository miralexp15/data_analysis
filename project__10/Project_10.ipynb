## 1. Загрузка и предобработка данных

```python
# <импорт необходимых библиотек>
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from scipy import stats as st
```

### 1.1. Загрузка данных и вывод общей информации

```python
# <загрузка данных в датафрейм>
try:
    df = pd.read_csv('bank_dataset.csv') # <запись файла csv в датафрейм для локальной машины>
except:
    df = pd.read_csv('/datasets/bank_dataset.csv') # <запись файла csv в датафрейм в онлайн-версии>

df.head(10) # <вывод первых десяти строк таблицы>
```

```python
# <вывод общей информации по датафрейму>
df.info()
```

Данные в столбцах:

- **userid** — идентификатор пользователя;
- **score** — баллы кредитного скоринга;
- **City** — город;
- **Gender** — пол;
- **Age** — возраст;
- **Objects** — количество объектов в собственности;
- **Balance** — баланс на счёте;
- **Products** — количество продуктов, которыми пользуется клиент;
- **CreditCard** — есть ли кредитная карта,
- **Loyalty** — активный клиент;
- **estimated_salary** — заработная плата клиента;
- **Churn** — ушёл или нет.

### 1.2. Смена названий столбцов

Часть названий столбцов начинается с заглавной буквы. Переведем все названия в нижний регистр, а затем для удобства разделим слова нижним подчеркиванием.

```python
# <смена названий столбцов>
df.columns = df.columns.str.lower()
df = df.rename(columns = {'userid':'user_id', 'creditcard':'credit_card'})
df.columns
```

Дополнительно поменяем порядок столбцов, отсортировав их по смыслу и типу данных для удобства работы. Ключевой для исследования столбец *products* перенесем в конец.

```python
# <смена порядка столбцов>
df = df[['user_id', 'city', 'gender', 'age', 'credit_card', 'loyalty', 
         'churn', 'objects', 'score', 'balance', 'estimated_salary', 'products']]
df.head()
```

## 2. Исследовательский анализ данных

Для автоматизации построения гистограмм сначала напишем функцию для их отрисовки.

```python
# <функция для построения гистограмм и столбчатых диаграмм>
def hist_plt(column, x_label, palette='Greys', ishist='n', n1=0, n2=0, n12=0):
    sns.set_style("whitegrid") # <установка стиля диаграмм>
    fig, ax = plt.subplots(1, 1, figsize=(15, 5)) # <подготовка контейнера для диаграммы>
    plt.title('Распределение значений столбца '+column, fontsize=20) # <название диаграммы>
    
    # <если ishist='n', строим barplot>
    if ishist == 'n':
        temp = df[column].value_counts().sort_index() # <создание переменной с данными колонки>
        sns.barplot(x=temp.index, y=temp, alpha=0.75, ax=ax, palette=palette) # <barplot>
        plt.xlabel(x_label, fontsize=16) # <подпись оси x>
        plt.ylabel('Абсолютное количество клиентов', fontsize=16) # <подпись оси y>
        plt.xticks(fontsize=14) # <настройка размеров шрифта xticks>
        plt.yticks(fontsize=14) # <настройка размеров шрифта yticks>
        return ax # <возврат ax для настройки xticks>
    
    # <если ishist='y', строим гистограмму>
    elif ishist == 'y':
        temp = df[column] # <создание переменной с данными колонки>
        # <n1, n2, n12 - задают диапазоны bins для упрощения восприятия данных>
        sns.distplot(temp, color='darkslategray', bins=np.arange(n1, n2, n12), kde=False, norm_hist=False)
        plt.xlabel(x_label, fontsize=16) # <подпись оси x>
        plt.ylabel('Абсолютное количество клиентов', fontsize=16) # <подпись оси y>
        plt.xticks(fontsize=14) # <настройка размеров шрифта xticks>
        plt.yticks(fontsize=14) # <настройка размеров шрифта yticks>
```

```python
# <функция для подсчета абсолютных и относительных значений количества клиентов>
def abs_and_rel(column):
    temp1 = df[column].value_counts().sort_index() # <подсчет абсолютных значений>
    temp2 = temp1 / temp1.sum() * 100 # <подсчет относительных значений в процентах>
    temp1.name = 'abs' # <переименование Series для абсолютных значений>
    temp2.name = 'rel (%)' # <переименование Series для относительных значений>
    return pd.DataFrame(temp1).join(temp2) # <вывод таблицы>
```

## 3. Сегментация пользователей на основе данных о количестве потребляемых продуктов

### 3.1. Определение количества групп

Перед построением гистограмм с разделением по количеству потребляемых продуктов необходимо проверить, будет ли корректно разделить выборку на тех, кто пользуется одним продуктом, и тех, кто пользуется несколькими.

Для этого сравним значение *churn* по группам. Отток - один из самых важных показателей.

```python
# <создание таблицы с квантилями churn с разделением по количеству продуктов>
temp = df[df['products'] == 1]['churn'].quantile(np.arange(.05, .96, .05)) # <расчет для products=1>
for i in [2, 3, 4]:
    # <добавление столбцов для products = 2, 3, 4>
    temp = pd.concat([temp, df[df['products'] == i]['churn'].quantile(np.arange(.05, .96, .05))], axis=1)

temp = temp.reset_index() # <перевод индексов в колонки>
temp.columns=['quantiles'] + [str(i)+' products' for i in range(1, 5)] # <смена названий столбцов>
temp.style.highlight_max(color='tomato', axis=None).highlight_min(color='lightgreen', axis=None)\
    .set_properties(padding='5px', border='1px solid black') # <выделение максимумов и минимумов и улучшение читаемости>
```

```python
# <проверка, все ли клиенты, пользующиеся 4мя продуктами уходят в отток>
df[df['products'] == 4]['churn'].value_counts()
```

Выводы по полученной таблице:

- реже всех уходят в отток клиенты, пользующиеся двумя продуктами;
- **все клиенты**, пользующиеся 4мя продуктами, *уходят в отток*;
- также высока доля оттока у клиентов, которые пользуются тремя продуктами.

Эти выводы являются контринтуитивными. На первый взгляд кажется, что если клиент пользуется большим количеством продуктов, то склонен к долгосрочному сотрудничеству. Однако данные не подтверждают этого.

Удалось выяснить, что делить на 2 группы (1 продукт и несколько продуктов) некорректно, так как один из ключевых показателей (*churn*) значительно отличается в разных группах.

Принимаем решение *провести сегментирование с разделением на 4 группы*: клиенты, пользующиеся *одним*, *двумя*, *тремя* и *четырьмя* продуктами.

### 3.2. Исследование признаков при различных значениях products

#### 3.2.1. Столбец city

```python
# <функция для подсчета относительных значений (для колонок с небольшим разнообразием значений)>
def rel_values(column):
    # <подсчет количества значений и деление на размер этой части выборки>
    # <подготовка таблицы с разбивкой по products>
    temp1 = (df.groupby('products')[column].value_counts() / df.groupby('products').size()).unstack()
    # <подготовка таблицы со значениями для всей таблицы>
    temp2 = pd.DataFrame(df[column].value_counts() / len(df))
    temp2.columns=['total']
    # <соединение датафреймов>
    return pd.concat([temp1, temp2.T], sort=True)
    
```

- В группе с количеством продуктов 3 меньше всего клиентов из Ярославля (*39%*) при средней доле Ярославля во всем датасете около 50%;

```python
# <функция для построения гистограмм для 4х значений products>
# <n и m вводятся для настройки количества bins>
def hist_plt_products(column, x_label, n=1, m=0):
    sns.set_style("whitegrid") # <установка стиля диаграмм>
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10)) # <создание subplots>
    plt.suptitle('Распределение значений столбца '+column, fontsize=20) # <название диаграммы>
    
    # <создание 4х графиков в цикле>
    for i, j in zip(range(1, 5), [ax1, ax2, ax3, ax4]):
        temp = df[df['products'] == i][column] # <создание временной таблицы>
        # <построение диаграммы и добавление вспомогательных элементов>
        sns.distplot(temp, color='darkslategray', kde=False, norm_hist=False, ax=j, bins=int((temp.max()-temp.min())/n)+m)
        j.axvline(temp.mean(), color='red', label='среднее = '+str(round(temp.mean(), 2)))
        j.axvline(temp.median(), color='blue', label='медиана = '+str(round(temp.median(), 2)))
        j.set_title('Для products = '+str(i))
        j.set_xlabel(x_label)
        j.set_ylabel('Абсолютное количество значений')
        j.legend()
    
    # <настройка одинакового масштаба для всех subplots>
    for j in [ax2, ax3, ax4]:
        ax1.get_shared_x_axes().join(ax1, j)
        j.autoscale()
        
    # <настройка отступов>
    fig.subplots_adjust(top=0.9)
```

## 4. Проверка статистических гипотез

### 4.1. Проверка гипотезы 1: «Между распределениями выборок доходов клиентов, пользующихся одним продуктом и пользующихся двумя продуктами, нет различий»

Подготовим данные и построим гистограммы распределений доходов клиентов.

```python
# <создание таблиц: клиенты с одним продуктом и клиенты с двумя продуктами>
sal_1_prod = df[df['products'] == 1]['estimated_salary']
sal_2_prod = df[df['products'] == 2]['estimated_salary']
```

```python
# <общая информация по значениям дохода для двух групп>
temp=pd.concat([sal_1_prod.describe(), sal_2_prod.describe()], axis=1)
temp.columns=['est_salary_1_product', 'est_salary_2_products']
temp
```

```python
# <построение гистограмм распределений для двух значений количества продуктов>
# <создание контейнера для графика>
plt.figure(figsize=(15, 5))
plt.title('Сравнение распределений доходов клиентов, пользующихся одним и двумя продуктами', fontsize=20)

# <график для пользователей одного продукта bins=20, каждый столбец представляет долю от общего количества>
sal_1_prod.hist(bins=np.arange(0, 200001, 10000), alpha=0.4, weights=np.ones_like(np.arange(len(sal_1_prod))) 
                / len(sal_1_prod))

# <график для пользователей двух продуктов bins=20, каждый столбец представляет долю от общего количества>
sal_2_prod.hist(bins=np.arange(0, 200001, 10000), alpha=0.4, weights=np.ones_like(np.arange(len(sal_2_prod))) 
                / len(sal_2_prod))

# <легенда и названия осей>
plt.legend(['Один продукт', 'Два продукта'])
plt.xlabel('Доход')
plt.ylabel('Доля значений');
```

Распределения значений для обеих групп не являются нормальными. Поэтому для проверки гипотез воспользуемся критерием Манна-Уитни.

**Нулевая гипотеза (H0)**: между распределениями выборок доходов клиентов, пользующихся *одним продуктом* и пользующихся *двумя продуктами*, нет различий;

**Альтернативная гипотеза (Н1)**: между распределениями выборок доходов клиентов, пользующихся *одним продуктом* и пользующихся *двумя продуктами*, есть различия.

```python
# <альфа - пороговый уровень статистической значимости>
alpha = 0.05
```

```python
# <проверка статистической гипотезы>
test = st.mannwhitneyu(sal_1_prod, sal_2_prod)[1]
print('p-value = {:.4f}'.format(test))
if test < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**P-value** равен **0.204**. Это означает, что:
1. С вероятностью около **20.4%** имеющиеся различия можно получить случайно;
2. P-value (**0.204**) больше значения альфа (**0.05**), поэтому не получается отклонить нулевую гипотезу.

*Между распределениями выборок доходов клиентов, пользующихся одним продуктом и пользующихся двумя продуктами, нет различий.*

```python
# <проверка распределений на нормальность>
# <альфа - пороговый уровень статистической значимости>
alpha = 0.05

# <первая выборка>
p_value = st.shapiro(score_1_prod)[1]
print('Нулевая гипотеза: Распределение кредитного скоринга для клиентов с одним продуктом нормальное')
print('Альтернативная гипотеза: Распределение кредитного скоринга для клиентов с одним продуктом ненормальное')
print()
print('p-значение для первой группы: ', p_value)
if p_value < alpha:
    print('Отвергаем нулевую гипотезу: распределение не нормальное')
else:
    print('Не получилось отвергнуть нулевую гипотезу, распределение нормальное')
print()
print()

# <вторая выборка>
p_value = st.shapiro(score_2_prod)[1]
print('Нулевая гипотеза: Распределение кредитного скоринга для клиентов с двумя продуктами нормальное')
print('Альтернативная гипотеза: Распределение кредитного скоринга для клиентов с двумя продуктами ненормальное')
print()
print('p-значение для второй группы: ', p_value)
if p_value < alpha:
    print('Отвергаем нулевую гипотезу: распределение не нормальное')
else:
    print('Не получилось отвергнуть нулевую гипотезу, распределение нормальное')
```

Распределения не являются нормальными, поэтому проводим непараметрический тест Манна-Уитни.

**Нулевая гипотеза (H0)**: между распределениями выборок кредитного скоринга клиентов, пользующихся *одним продуктом* и пользующихся *двумя продуктами*, нет различий;

**Альтернативная гипотеза (Н1)**: между распределениями выборок кредитного скоринга клиентов, пользующихся *одним продуктом* и пользующихся *двумя продуктами*, есть различия.

```python
# <альфа - пороговый уровень статистической значимости>
alpha = 0.05
```

```python
# <проверка статистической гипотезы>
test = st.mannwhitneyu(score_1_prod, score_2_prod)[1]
print('p-value = {:.4f}'.format(test))
if test < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**P-value** равен **0.0786**. Это означает, что:
1. С вероятностью около **7.86%** имеющиеся различия можно получить случайно;
2. P-value (**0.0786**) больше значения альфа (**0.05**), поэтому не получается отклонить нулевую гипотезу.

*Между распределениями выборок кредитного скоринга клиентов, пользующихся одним продуктом и пользующихся двумя продуктами, нет различий.*
